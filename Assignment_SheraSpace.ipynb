```python
#Reading the pdf and extracting the text
from tika import parser

raw = parser.from_file('/content/Resume - Shad Al Kaiser.pdf')
print(raw['content'])
```

```python
#converting every alphabet from the pdf into lowercase for preprocessing purpose
text=raw['content']
corpus=text.lower()
```

```python
#importing nltk for tokenizing the text corpus
import nltk
nltk.download('punkt')
from nltk.tokenize import word_tokenize
```

```python
#tokenizing and removing the duplicate tokens from the set
token = word_tokenize(corpus)
token = list(dict.fromkeys(token))
token
```

```python
#given skill dataset by the company

skills = ['python', 'react', 'html', 'css', 'java']
```

```python
#counting the number of matching skills and printing the verdict based on the final result
count = 0
for i in token:
  for j in skills:
    if i==j:
        count = count+1
        print(i)
if count==0:
  print("Very Bad skill match")
if count>0 and count<3:
  print("Good skill match")
if count>3:
  print("Excellent skill match")
```
